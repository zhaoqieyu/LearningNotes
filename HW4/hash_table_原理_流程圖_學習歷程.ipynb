```python
from Crypto.Hash import MD5
class ListNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class MyHashSet:
    def __init__(self, capacity=5):
        self.capacity = capacity
        self.data = [None] * capacity
        
    def MD5_hash(self,data):
        a = MD5.new()
        a.update(data.encode("utf-8"))
        x = a.hexdigest()
        index = int(x,16)%self.capacity
        return index
           
    def add(self, key):
        index = self.MD5_hash(key)
        if self.data[index] == None: #è‹¥åŸæœ¬è©²bucketsä¸­æ²’æœ‰å€¼æ™‚
            self.data[index] = ListNode(key) #å‰‡æŠŠkeyç›´æ¥åŠ å…¥è©²indexä¸­
            return
        else: #è‹¥åŸæœ¬è©²bucketsä¸­å·²ç¶“æœ‰å€¼æ™‚
            node = self.data[index] #æ‰¾åˆ°è¦æŠŠkeyæ”¾åœ¨å“ªå€‹bucketsï¼Œä¸¦å­˜æ”¾åœ¨nodeä¸­
            while node.next != None : #çœ‹è©²nodeçš„ä¸‹ä¸€å€‹(å¾€è©²bucketå¾Œé¢çœ‹)æ˜¯å¦ç‚ºç©ºï¼Œè‹¥ä¸æ˜¯ç©ºçš„
                node = node.next #å…ˆæŠŠç›®å‰çš„nodeæŒ‡å‘ä¸‹ä¸€å€‹node
            node.next = ListNode(key) #å†æŠŠkeyç›´æ¥åŠ å…¥ä¸‹ä¸€å€‹nodeä¸­
        
    def remove(self, key):
        index = self.MD5_hash(key)
        if self.data[index] != None and self.data[index].data == key: #è‹¥åŸæœ¬è©²buckets(index)éç‚ºç©ºï¼Œä¸”è©²bucket(index)ä¸­çš„å€¼å³ç‚ºåˆªé™¤å€¼
            if self.data[index].next == None: #è‹¥æ­¤åˆªé™¤å€¼æ˜¯Linked Listä¸­çš„æœ€å¾Œä¸€å€‹(ä¸‹ä¸€å€‹å€¼ç‚ºç©º)
                self.data[index] = None #ç›´æ¥æŠŠæœ€å¾Œä¸€å€‹åˆªé™¤(æ”¹æˆç©º)
            else: #è‹¥æ­¤åˆªé™¤å€¼ä¸æ˜¯Linked Listä¸­æœ€å¾Œä¸€å€‹
                self.data[index] = self.data[index].next #å°‡è¦åˆªé™¤çš„å€¼çš„å‰ä¸€å€‹indexç›´æ¥æŒ‡å‘è¦åˆªé™¤çš„å€¼çš„ä¸‹ä¸€å€‹index
            return True
        elif ( self.data[index] != None and self.data[index].next != None ): #è‹¥åŸæœ¬è©²buckets(index)éç‚ºç©ºï¼Œä¸”è©²bucket(index)çš„ä¸‹ä¸€å€‹éç‚ºç©º
            next_node = self.data[index].next #æŠŠè©²bucket(index)çš„ä¸‹ä¸€å€‹å­˜æ”¾ç‚ºnext_node
            while next_node != None : #ç•¶next_nodeä¸ç‚º0
                if  next_node.data == key: #ä¸”next_nodeç­‰æ–¼æœå°‹å€¼
                    if next_node.next == None :#è‹¥æ­¤åˆªé™¤å€¼æ˜¯Linked Listä¸­çš„æœ€å¾Œä¸€å€‹(ä¸‹ä¸€å€‹å€¼ç‚ºç©º)
                        next_node = None #ç›´æ¥æŠŠæœ€å¾Œä¸€å€‹åˆªé™¤(æ”¹æˆç©º)
                    else: #è‹¥æ­¤åˆªé™¤å€¼ä¸æ˜¯Linked Listä¸­æœ€å¾Œä¸€å€‹
                        next_node =next_node.next #å¾€ä¸‹ä¸€å€‹ç¹¼çºŒ
                    return True
                else: #ä¸”next_nodeä¸ç­‰æ–¼æœå°‹å€¼
                    next_node = next_node.next #å¾€ä¸‹ä¸€å€‹ç¹¼çºŒ
        return False
  
    def contains(self, key):
        index = self.MD5_hash(key)
        if self.data[index] != None and self.data[index].data == key: #è‹¥åŸæœ¬è©²buckets(index)éç‚ºç©ºï¼Œä¸”è©²bucket(index)ä¸­çš„å€¼å³ç‚ºæœå°‹å€¼
            return True #ä»£è¡¨æ‰¾åˆ°äº†ï¼Œå›å‚³True
        elif ( self.data[index] != None and self.data[index].next != None ): #è‹¥åŸæœ¬è©²buckets(index)éç‚ºç©ºï¼Œä¸”è©²bucket(index)çš„ä¸‹ä¸€å€‹éç‚ºç©º
            next_node = self.data[index].next #æŠŠè©²bucket(index)çš„ä¸‹ä¸€å€‹å­˜æ”¾ç‚ºnext_node
            while next_node != None: #ç•¶next_nodeä¸ç‚º0
                if  next_node.data == key: #ä¸”next_nodeç­‰æ–¼æœå°‹å€¼
                    return True #ä»£è¡¨æ‰¾åˆ°äº†ï¼Œå›å‚³True
                else: #è‹¥next_nodeä¸ç­‰æ–¼æœå°‹å€¼
                    next_node = next_node.next #å†ç¹¼çºŒå¾€ä¸‹æ‰¾(next_node.next)
        return False
```

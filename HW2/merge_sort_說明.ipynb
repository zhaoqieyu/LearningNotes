```python
def split(example):                             
    middle = len(example)//2+1                  #åˆ†å‰²æˆå…©éƒ¨åˆ†(å¾ä¸­é–“åˆ†)
    return example[:middle],example[middle:]    #å›å‚³ç¬¬0ç­†åˆ°middleä»¥åŠå›å‚³middleä»¥å¾Œçš„å€¼
split(example)
```

æ­£ç¢ºçš„æ‡‰è©²æ˜¯è¦åˆ†å‰²æˆåªå‰©ä¸‹ä¸€å€‹å…ƒç´ ï¼Œæœ¬ä¾†ä»¥ç‚ºæˆ‘æ‰“çš„é€™ç¨®å°±å¯ä»¥äº†ï¼Œä½†çµæœé¡¯ç¤ºåªåˆ†å‰²äº†ä¸€æ¬¡ï¼Œç„¡æ³•åˆ†å¾ˆå¤šæ¬¡ï¼Œç™¼ç¾å¿˜äº†åšè¿´åœˆ!!

```python
def split2(example):
    
    if len(example)<=1:                    #è‹¥æ•¸åˆ—é•·åº¦<=1
        return(example)                    #ç›´æ¥å›å‚³æ•¸åˆ—
    
    else:
        middle = len(example)//2+1         #è¨­:è¦å¾å“ªè£¡åˆ†
    
        if len(example[:middle])<=1:       #è‹¥exampleå‰åŠæ®µ<=1
            return example[:middle]        #ä¸ç”¨åˆ†å‰²ï¼Œç›´æ¥å›å‚³å‰åŠæ®µ
        else:
            split2(example)                #å¦å‰‡ï¼Œåˆ†å‰²example
            
        if len(example[middle:])<=1:       #è‹¥exampleå¾ŒåŠæ®µ<=1
            return example[middle:]        #ä¸ç”¨åˆ†å‰²ï¼Œç›´æ¥å›å‚³å¾ŒåŠæ®µ
        else:
            split2(example)                #å¦å‰‡ï¼Œåˆ†å‰²example  
            
    return example[:middle],example[middle:]#æœ€å¾Œç›´æ¥å›å‚³å‰åŠåŠ å¾ŒåŠ

print(split2(example))          
```

```python
def split3(example):
    print(example)
    middle = len(example)//2
    
    if len(example)>1:           
        return tuple(split3(example[:middle]))+tuple(split3(example[middle:]))      
    elif len(example)==1:                                                           
        return [example]
    
split3(example) 
```

```python
def merge_sorted_array(left_array,right_array):                   
    if len(left_array) == 0:                  #è‹¥å·¦å³å…©é‚Šä¸­ä¸€å€‹æˆ–å…©å€‹éƒ½=0
        return left_array
    elif len(right_array) == 0:
        return left_array
    else:    
        new_array=[]
        #å»ºç«‹new_array,ä»¥ä¾¿æŠŠleft_arrayå’Œright_arrayçš„å€¼æ”¾å…¥
        x=0               #è¨­index:left_arrayçš„indexå–åç‚ºx
        y=0               #è¨­index:right_arrayçš„indexå–åç‚ºy
        i=0               #è¨­index:new_arrayçš„indexå–åç‚ºi
    
    while:
        if left_array[x] < right_array[y]:    #è‹¥xä¸­çš„å€¼ < yä¸­çš„å€¼
            new_array[i] = left_array[x]      #æŠŠxä¸­çš„å€¼æ”¾å…¥new_array
            x+=                               
            i+=
            #new_arrayçš„ä¸‹ä¸€ç­†å€¼(i+1)è®Šæˆ :
            #left_arrayçš„ä¸‹ä¸€ç­†è³‡æ–™(x+1)å’Œright_arrayåŸæœ¬çš„å€¼(y)åšæ¯”è¼ƒ
            
        else(left_array[x] >= right_array[y]):
            new_array[i] = right_array[y]
            y+=
            i+=
        
        if x == len(left_array):              #ç•¶left_arrayå·²ç¶“æ²’æœ‰å€¼çš„æ™‚å€™                                    
            new_array.append(right_array[y:]) 
            #ä¸ç”¨å†åšæ¯”è¼ƒ,ç›´æ¥æŠŠright_arrayå‰©ä¸‹çš„å€¼åŠ å…¥new_arrayå³å¯
            break                             #è¿”å›ä¸¦é€€å‡ºæ•´å€‹å¾ªç’°
        else y== len(right_array):
            new_array.append(left_array[x:])
            break
    return new_array
```

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Nov  3 18:35:01 2019

@author: jessica
"""


class Solution(object):

    def merge_sorted_array(self, left_array, right_array):   #æ¯”å¤§å°
        print('left_array')
        print(left_array)
        print('right_array')
        print(right_array)
        
        if len(left_array) == 0:               #è‹¥å·¦é‚Šçš„é•·åº¦=0
            return left_array                  #ç›´æ¥å›å‚³å·¦é‚Š
        elif len(right_array) == 0:            #è‹¥å³é‚Šçš„é•·åº¦=0
            return right_array                 #ç›´æ¥å›å‚³å·¦é‚Š
        else:                                  #å¦å‰‡ï¼Œåœ¨å·¦å³å…©é‚Šé•·åº¦çš† >0çš„æƒ…æ³ä¸‹
            new_array = []                     #å»ºç«‹new_arrayï¼Œä»¥ä¾¿æŠŠleft_arrayå’Œright_arrayçš„å€¼æ”¾å…¥    
            x=0                                #è¨­index:left_arrayçš„indexå–åç‚ºx
            y=0                                #è¨­index:right_arrayçš„indexå–åç‚ºy
            i=0                                #è¨­index:new_arrayçš„indexå–åç‚ºi

            if left_array[x] < right_array[y]:      #è‹¥xä¸­çš„å€¼ < yä¸­çš„å€¼
                new_array.append(left_array[x])     #æŠŠxä¸­çš„å€¼æ”¾å…¥new_array
                del left_array[x]                   
                #ä¸¦åˆªæ‰åŸæœ¬åœ¨left_array[x]ä¸­çš„å€¼ï¼Œä»¥ä¾¿å¾ŒçºŒå±¬é•·åº¦æ™‚éƒ¨æœƒå‡ºéŒ¯
                x += x
                i += i
                #new_arrayçš„ä¸‹ä¸€ç­†å€¼(i+1)è®Šæˆ :
                #left_arrayçš„ä¸‹ä¸€ç­†è³‡æ–™(x+1)å’Œright_arrayåŸæœ¬çš„å€¼(y)åšæ¯”è¼ƒ
            else:                                   #åŒç†
                new_array.append(right_array[y])
                del right_array[y]
                y += y
                i += i

            if len(left_array) == 0:            #æœ€å¾Œï¼Œç•¶å·¦é‚Šé•·åº¦=0æ™‚ï¼Œå³å·¦é‚Šå·²ç¶“æ¯”å®Œï¼Œæ²’æœ‰å€¼äº†
                new_array += right_array        #å‰‡ç›´æ¥æŠŠå³é‚Šå‰©ä¸‹çš„å…¨éƒ¨åŠ åˆ°new_arrayå…§
            elif len(right_array) == 0:         #åŒç†
                new_array += left_array             
            else:
                new_array.extend(self.merge_sorted_array(left_array, right_array))   
                #ç”¨selfå‘¼å«æŠŠå·¦å³å…©é‚Šé€émerge_sorted_arrayåšæ¯”è¼ƒï¼Œ
                #å†ç”¨extendæŠŠå·¦å³å…©é‚Šæ¯”è¼ƒéå¾Œçš„åŠ å…¥new_arrayçš„å°¾ç«¯
        print('return')
        print(new_array)
        return new_array

    
    def merge_sort(self, example):                  #ä¸»ç¨‹å¼:åˆ†å‰²å¾Œä¸¦åšåˆä½µæ’åº
        #print(example)
        middle = len(example)//2                    #åˆ†å‰²æˆå…©éƒ¨åˆ†(å¾ä¸­é–“åˆ†)

        if len(example) > 1: 
            #ç•¶å·²ç¶“æ‹†åˆ°åº•æ™‚ï¼Œä¹‹å¾Œçš„åˆä½µ
            return self.merge_sorted_array(self.merge_sort(example[:middle])
            #åŸæœ¬å¯«çš„:return tuple(split2(example[:middle]))+tuple(split2(example[middle:]))
        else:                     
            return example


example=[9,6,21,19,23,8,6]

output = Solution().merge_sort(example)
print(output)
```

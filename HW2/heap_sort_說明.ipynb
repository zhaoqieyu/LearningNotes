```python
def Change(self, i, j): #äº¤æ›çš„æ–¹æ³•
        self.list[j], self.list[i] = self.list[i], self.list[j] 
    
    def Compare(self, firstIndex, secondIndex): #æ¯”è¼ƒå¾Œäº¤æ›çš„é‚è¼¯
    #first_Indexå‚³é€²ä¾†ç¬¬ä¸€å€‹å€¼ã€second_Indexå‚³é€²ä¾†ç¬¬äºŒå€‹å€¼
        if(self.maxOrmin == "max"): #max heap
            return self.list[firstIndex] < self.list[secondIndex]
        elif(self.maxOrmin == "min"): #min heap
            return self.list[firstIndex] > self.list[secondIndex]
        
    def IsExistIndex(self, index): #ç¢ºä¿indexå°æ–¼æœ€å¾Œä¸€å€‹çµé»ï¼Œä¹Ÿå°±æ˜¯èªªåœ¨ç¯„åœå…§ï¼Œä¸æœƒè¶…å‡ºç¯„åœ
        return index < self.changeEnd 

    def CompareResult(self, maxIndex, compareIndex):
        if self.IsExistIndex(compareIndex) and self.Compare(maxIndex, compareIndex):  
        #æœ‰å…©å€‹æ¢ä»¶(ç”¨å‘¼å«çš„æ–¹å¼)
            maxIndex = compareIndex #ç¬¦åˆçš„è©±å³ç‚ºæœ€å¤§å€¼
        return maxIndex
```

```python
class Solution(object):
    list = []
    maxOrmin = "max"  #max heap
    changeEnd = 0 #è¨˜éŒ„è¦åšçš„æœ€å¾Œä¸€å€‹çµé»
    
    def Heapify(self, currentIndex): #çœ‹çœ‹ç•¶å‰çš„çˆ¶ç¯€é»å’Œä¸‹é¢çš„å·¦æœ‰å…©å€‹ç¯€é»(å…±3å€‹å€¼)

        left_child_Index = self.LeftChildIndex(currentIndex) 
          #ç”¨ç•¶å‰çš„indexå¸¶å…¥å…¬å¼ï¼Œå»å‘¼å«ï¼Œå¯ä»¥æ¨çŸ¥ç•¶å‰çš„é€™å€‹ç¯€é»çš„å·¦é‚Šçš„index
        right_child_Index = self.RightChildIndex(currentIndex) 
          #ç”¨ç•¶å‰çš„indexå¸¶å…¥å…¬å¼ï¼Œå»å‘¼å«ï¼Œå¯ä»¥æ¨çŸ¥ç•¶å‰çš„é€™å€‹ç¯€é»çš„å³é‚Šçš„index
        maxIndex = currentIndex 
          #å…ˆå‡è¨­ç•¶å‰çš„çˆ¶ç¯€é»æ˜¯3å€‹é˜ä¸­çš„æœ€å¤§å€¼

        #æ‰¾å‡ºæœ€å¤§å€¼
        maxIndex = self.CompareResult(maxIndex, left_child_Index) #å·¦é‚Šå’Œæœ€å¤§å€¼åšæ¯”è¼ƒ
        maxIndex = self.CompareResult(maxIndex, right_child_Index) #å³é‚Šå’Œæœ€å¤§å€¼åšæ¯”è¼ƒ

        #å†æ‹¿æœ€å¤§å€¼å’Œiåšäº¤æ›
        if maxIndex != currentIndex:  
        #é‚£ä½•æ™‚ä¸ç”¨åšäº¤æ›?åœ¨ç•¶å‰çš„çˆ¶ç¯€é»å·²æ˜¯æœ€å¤§å€¼æ™‚ï¼Œå› æ­¤åœ¨ç•¶å‰çš„çˆ¶ç¯€é»ä¸ç­‰æ–¼æ˜¯æœ€å¤§å€¼æ™‚ 
            self.Change(currentIndex, maxIndex) #äº¤æ›
            return self.Heapify(maxIndex) #äº¤æ›å®Œå¾Œå†å°å®ƒçš„åº•å±¤çš„é‚£ä¸€å †å†åšä¸€æ¬¡heapify
    
    def heap_sort(self,list):
        self.list=list
        rootIndex = 0 #é ‚é»
        endIndex = len(self.list) #å¾æœ€å¾Œä¸€å€‹é–‹å•Ÿ
        startIndex = endIndex // 2 - 1 
          #é–‹å§‹åŸ·è¡Œçµ„åˆ(è©²å †)æœ€å¤§æ•¸çš„å•Ÿé»
          #å¥—å…¬å¼ç®—ä¸€ç®—ï¼Œæ˜¯ç‚ºäº†æ’é™¤æœ€å¾Œä¸€å€‹è‘‰å­çš„ç¯€é»ï¼Œå› ç‚ºæœ€å¾Œä¸€å€‹ç¯€é»ä¸ç”¨æ’
        
        #æŠŠæœ€å¤§çš„æ•¸å€¼æ¨åˆ°0çš„ä½ç½®ï¼Œæ¯ä¸€éšæ®µéƒ½è·Ÿå…©å€‹å­ç¯€é»æ¯”è¼ƒï¼Œæœ€å¤§çš„å¾€ä¸Šç§»å‹•ï¼Œæœ€å¾Œåœ¨é ‚é»çš„å°±æ˜¯æœ€å¤§
        for index in range(startIndex, -1, -1):
            self.changeEnd = endIndex
            self.Heapify(index)
        
        #é ‚é»èˆ‡æœ€å¾Œæ•¸å€¼çš„äº¤æ›ï¼Œäº¤æ›å®Œå¾Œå†é‡æ–°æ’ä¸€æ¬¡
        for index in range(endIndex-1, 0, -1): #endIndexï¼šå¾æœ€å¾Œä¸€å€‹é–‹å•Ÿï¼Œéš¨è‘—æ¯ä¸€å€‹å›åˆæ¸›1
            self.changeEnd = index
            self.Change(index, rootIndex)
            self.Heapify(rootIndex)
            
   
    def LeftChildIndex(self, parent_Index):
        return 2*parent_Index + 1 #é€éçˆ¶ç¯€é»æ¨çŸ¥å·¦é‚Šçš„å­ç¯€é»
    
    def RightChildIndex(self, parent_Index):
        return 2*parent_Index + 2 #é€éçˆ¶ç¯€é»æ¨çŸ¥å³é‚Šçš„å­ç¯€é»

    
    def Change(self, i, j): #äº¤æ›çš„æ–¹æ³•
        self.list[j], self.list[i] = self.list[i], self.list[j] 
    
    def Compare(self, firstIndex, secondIndex): #æ¯”è¼ƒå¾Œäº¤æ›çš„é‚è¼¯
    #first_Indexå‚³é€²ä¾†ç¬¬ä¸€å€‹å€¼ã€second_Indexå‚³é€²ä¾†ç¬¬äºŒå€‹å€¼
        if(self.maxOrmin == "max"): #max heap
            return self.list[firstIndex] < self.list[secondIndex]
        elif(self.maxOrmin == "min"): #min heap
            return self.list[firstIndex] > self.list[secondIndex]
        
    def IsExistIndex(self, index): #ç¢ºä¿indexå°æ–¼æœ€å¾Œä¸€å€‹çµé»ï¼Œä¹Ÿå°±æ˜¯èªªåœ¨ç¯„åœå…§ï¼Œä¸æœƒè¶…å‡ºç¯„åœ
        return index < self.changeEnd 

    def CompareResult(self, maxIndex, compareIndex):
        if self.IsExistIndex(compareIndex) and self.Compare(maxIndex, compareIndex):  
        #æœ‰å…©å€‹æ¢ä»¶(ç”¨å‘¼å«çš„æ–¹å¼)
            maxIndex = compareIndex #ç¬¦åˆçš„è©±å³ç‚ºæœ€å¤§å€¼
        return maxIndex
    
list = [9,6,21,19,23,8,6]
print("beg--------->")
print(list)
Solution().heap_sort(list)
print("end--------->")
print(list)
```

```python
class Graph(): 

    def __init__(self, vertices):         
        self.V = vertices 
        self.graph = [] 
    
    def addEdge(self,u,v,w): 
        self.graph.append([u,v,w])
        
    check = lambda self, parent , i : i if parent[i] == -1 else self.check(parent,parent[i]) 
    
    
    def Kruskal(self):
        result_dic = {} # to store MST result
        index, output = 0, 0 # index:è™•ç†çš„é‚Š/output:å·²è¼¸å‡º(æ‰¾åˆ°)çš„é‚Š 
        self.graph = sorted(self.graph,key=lambda item: item[2]) # ä¾ç…§æ¬Šé‡æ‰¾å‡ºæœ€å°çš„é‚Š(æ’åº)
        parent = [-1]*self.V # å…ˆå‡è¨­parentç‚º-1
        
        while output < self.V -1 : #è‹¥å·²æ‰¾åˆ°çš„é‚Šçš„æ•¸é‡å°æ–¼å·²æ‰¾åˆ°çš„é‚Šçš„æ•¸é‡          
            u,v,w = self.graph[index] #å…ˆè®€å‡ºlistä¸­çš„3å€‹å…ƒç´ 
            index += 1 #å·²è™•ç†ç¬¬ä¸€å€‹é‚Šï¼Œç¹¼çºŒå¾€ä¸‹æ‰¾
            
            if self.check(parent, u) !=  self.check(parent ,v) : #åˆ¤æ–·æ˜¯å¦æœ‰cicle(çœ‹u,vçš„parentæ˜¯å¦ç›¸åŒ)ï¼Œè‹¥ä¸åŒï¼Œä»£è¡¨æ²’æœ‰cicle 
                output += 1 #å‰‡æŠŠè©²é»åŠ å…¥output
                result_dic[str(u)+"-"+str(v)]=w #u,vç¢ºèªæ‰¾åˆ°ï¼ŒåŠ å…¥dictionary
                parent[self.check(parent, self.check(parent, u))] = self.check(parent, self.check(parent, v)) #æ›´æ”¹parentçš„å€¼ 
        return result_dic
```
